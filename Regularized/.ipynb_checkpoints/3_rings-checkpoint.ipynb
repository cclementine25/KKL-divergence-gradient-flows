```python
def generate_rings(m, radius, separation):
    points = []
    
    # Définition des centres des anneaux
    centers = [(0, 0), (0, separation), (0, -separation)]
    
    # Génération des points pour chaque anneau
    for center in centers:
        theta = np.linspace(0, 2*np.pi, m)
        y = center[0] + radius * np.cos(theta)
        x = center[1] + radius * np.sin(theta)
        ring_points = np.column_stack((x, y))
        points.extend(ring_points)
    
    return np.array(points)



def initialization(n, radius, separation, variance):
    # Définition du centre du premier anneau (le plus à gauche)
    center = (-separation, 0)
    
    # Décalage de l'extrémité gauche du premier anneau
    x_left_end = center[0] - radius
    y_left_end = center[1]
    
    # Génération des points à partir d'une distribution normale centrée sur l'extrémité gauche du premier anneau
    x = np.random.normal(x_left_end, variance, n)
    y = np.random.normal(y_left_end, variance, n)
    
    return np.column_stack((x, y))
```

```python
def plot_points(points):
    plt.figure(figsize=(8, 6))
    plt.scatter(points[:, 0], points[:, 1], s=5, c='blue', marker='o')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Distribution des points dans R^2 formant 3 anneaux alignés horizontalement')
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
def _generate_data():
    N, r, _delta = 80, 0.3, 0.5
    
    X = np.c_[r * np.cos(np.linspace(0, 2 * np.pi, N + 1)), r * np.sin(np.linspace(0, 2 * np.pi, N + 1))][:-1]  # noqa
    for i in [1, 2]:
        X = np.r_[X, X[:N, :]-i*np.array([0, (2 + _delta) * r])]

    rs = np.random.RandomState(42)
    Y = rs.randn(N*(2+1), 2) / 100 - np.array([r/np.sqrt(2), r/np.sqrt(2)])
    Y = rs.randn(N*(2+1), 2) / 100 - np.array([0, r])

    Y = torch.from_numpy(Y).float()
    X = torch.from_numpy(X).float()
    Y.requires_grad = True
    return X, Y
```
